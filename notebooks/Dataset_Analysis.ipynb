```python
## Categorical columns... 
categorical_columns = [columns for columns in df.columns if len(df[columns].unique()) <= 10]
categorical_columns
```

```python
## Numerical columns...
numerical_columns = [columns for columns in df.columns if columns not in categorical_columns]
numerical_columns
```

#### Exploratory Data Analysis

```python
plt.figure(figsize=(30, 30))
sns.heatmap(df.corr(), annot=True)
```

```python
correlations = df.corr(numeric_only=True)['fetal_health'].drop('fetal_health')  # drop target itself
```

```python
correlations_sorted = correlations.sort_values()
print(correlations_sorted)
```

```python
cor_df = pd.DataFrame({'Feature': correlations_sorted.index, 'Correlation': correlations_sorted.values})
```

```python
# Plot boxplot of correlations
plt.figure(figsize=(50, 50))
sns.barplot(x=cor_df['Feature'], y=cor_df['Correlation'])
plt.title('Boxplot of Feature Correlations with Fetal Health (Sorted)')
plt.xlabel('Correlation')
plt.ylabel('Features')
plt.grid(True)
plt.show()
```
